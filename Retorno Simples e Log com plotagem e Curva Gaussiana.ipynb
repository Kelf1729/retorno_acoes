```python
import numpy as np #Importação das bibliotecas 
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
Itau = wb.DataReader('ITSA4.SA', data_source='yahoo', start='1995-1-1') #Captando daddos 
```

# Retorno Simples 
$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
Itau['retorno_simples'] = (Itau['Adj Close'] / Itau['Adj Close'].shift(1)) - 1 #calculando o retorno simples das ações 
print (Itau['retorno_simples']) #Visualização dos dados
```

# Retorno diário (Retorno Simples) 

```python
retorno_diario_s = Itau['retorno_simples'].mean()
retorno_diario_s
```

```python
retorno_anual_s = retorno_diario_s*250 
retorno_anual_s
```

# Retorno Log 
\begin{align}
ln (\frac{P_{t}}{P_{t-1}})
\end{align}

```python
Itau['retorno_log'] = np.log(Itau['Adj Close'] / Itau['Adj Close'].shift(1)) #retorno Log da ação
print (Itau['retorno_log'])
```

```python
retorno_diario_l = Itau['retorno_log'].mean() #retorno log diário da ação 
retorno_diario_l
```

```python
retorno_anual_l = retorno_diario_l*250
retorno_anual_l
```

# Plotagem de Dados (Retorno Simples e Retorno Log) 

```python
plt.subplot(2,1,1)
Itau['retorno_simples'].plot(figsize=(10, 5), color = 'red')
plt.title('Demonstração de Retorno Simples Ações Itaú')
plt.xlabel('Período de retorno')
plt.ylabel('Retorno do período')
plt.tight_layout()
plt.subplot(2,1,2)
Itau['retorno_log'].plot(figsize=(10, 5))
plt.title('Demonstração de Retorno Log Ações Itaú')
plt.xlabel('Período de retorno')
plt.ylabel('Retorno do período')
plt.tight_layout() #Ajuste para que o texto do gráfico não entre nos demais 
plt.show()
```

# Plotagem da curva Gaussiana de Retornos 

```python
histograma_simplesretorno = plt.hist(Itau['retorno_simples'], bins = 1000, align = 'mid', histtype = 'bar')
plt.title('Curva Gaussiana de Retorno Simples')
plt.xlabel('Retorno')
plt.ylabel('Frequência do retorno')
plt.xlim(-0.3 , 0.2)
plt.ylim(0, 100)
plt.grid(True)
```

```python
histograma_simplesretorno = plt.hist(Itau['retorno_log'], bins = 1000, align = 'mid', histtype = 'bar', color = 'red')
plt.title('Curva Gaussiana de Retorno log')
plt.xlabel('Retorno')
plt.ylabel('Frequência do retorno')
plt.xlim(-0.3 , 0.2)
plt.ylim(0, 90)
plt.grid(True)
```
